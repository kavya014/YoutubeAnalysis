```python
import pandas as pd
import json
```

```python
data_CA = pd.read_csv('CAvideos.csv')
data_DE = pd.read_csv('DEvideos.csv')
data_FR = pd.read_csv('FRvideos.csv')
data_GB = pd.read_csv('GBvideos.csv')
data_IN = pd.read_csv('INvideos.csv')
data_JP = pd.read_csv('JPvideos.csv',encoding='latin1')
data_KR = pd.read_csv('KRvideos.csv',encoding='latin1')
data_MX = pd.read_csv('MXvideos.csv',encoding='latin1')
data_RU = pd.read_csv('RUvideos.csv',encoding='latin1')
data_US = pd.read_csv('USvideos.csv')
```

```python
# creates a dictionary that maps `category_id` to `category`
CA_id_to_category = {}
DE_id_to_category = {}
FR_id_to_category = {}
GB_id_to_category = {}
IN_id_to_category = {}
JP_id_to_category = {}
KR_id_to_category = {}
MX_id_to_category = {}
RU_id_to_category = {}
US_id_to_category = {}

with open('CA_category_id.json', 'r') as f:
    data = json.load(f)
    for category in data['items']:
        CA_id_to_category[category['id']] = category['snippet']['title']
with open('DE_category_id.json', 'r') as f:
    data = json.load(f)
    for category in data['items']:
        DE_id_to_category[category['id']] = category['snippet']['title']
with open('FR_category_id.json', 'r') as f:
    data = json.load(f)
    for category in data['items']:
        FR_id_to_category[category['id']] = category['snippet']['title']
with open('GB_category_id.json', 'r') as f:
    data = json.load(f)
    for category in data['items']:
        GB_id_to_category[category['id']] = category['snippet']['title']
with open('IN_category_id.json', 'r') as f:
    data = json.load(f)
    for category in data['items']:
        IN_id_to_category[category['id']] = category['snippet']['title']
with open('JP_category_id.json', 'r') as f:
    data = json.load(f)
    for category in data['items']:
        JP_id_to_category[category['id']] = category['snippet']['title']
with open('KR_category_id.json', 'r') as f:
    data = json.load(f)
    for category in data['items']:
        KR_id_to_category[category['id']] = category['snippet']['title']
with open('MX_category_id.json', 'r') as f:
    data = json.load(f)
    for category in data['items']:
        MX_id_to_category[category['id']] = category['snippet']['title']
with open('RU_category_id.json', 'r') as f:
    data = json.load(f)
    for category in data['items']:
        RU_id_to_category[category['id']] = category['snippet']['title']
with open('US_category_id.json', 'r') as f:
    data = json.load(f)
    for category in data['items']:
        US_id_to_category[category['id']] = category['snippet']['title']
```

```python
type_str_list = ['category_id']
for column in type_str_list:
    data_CA[column] = data_CA[column].astype(str)
    data_DE[column] = data_DE[column].astype(str)
    data_FR[column] = data_FR[column].astype(str)
    data_GB[column] = data_GB[column].astype(str)
    data_IN[column] = data_IN[column].astype(str)
    data_JP[column] = data_JP[column].astype(str)
    data_KR[column] = data_KR[column].astype(str)
    data_MX[column] = data_MX[column].astype(str)
    data_RU[column] = data_RU[column].astype(str)
    data_US[column] = data_US[column].astype(str)
```

```python
data_CA['category_name'] = data_CA['category_id'].map(CA_id_to_category)
data_DE['category_name'] = data_DE['category_id'].map(DE_id_to_category)
data_FR['category_name'] = data_FR['category_id'].map(FR_id_to_category)
data_GB['category_name'] = data_GB['category_id'].map(GB_id_to_category)
data_IN['category_name'] = data_IN['category_id'].map(IN_id_to_category)
data_JP['category_name'] = data_JP['category_id'].map(JP_id_to_category)
data_KR['category_name'] = data_KR['category_id'].map(KR_id_to_category)
data_MX['category_name'] = data_MX['category_id'].map(MX_id_to_category)
data_RU['category_name'] = data_RU['category_id'].map(RU_id_to_category)
data_US['category_name'] = data_US['category_id'].map(US_id_to_category)
```

```python
data_CA['region']='Canada'
data_DE['region']='Denmark'
data_FR['region']='France'
data_GB['region']='Great Britan'
data_IN['region']='India'
data_JP['region']='Japan'
data_KR['region']='South Korea'
data_MX['region']='Mexico'
data_RU['region']='Russia'
data_US['region']='United States'
df=pd.DataFrame(pd.concat([data_CA,data_DE,data_FR,data_GB,data_IN,data_JP,data_KR,data_MX,data_RU,data_US], ignore_index=True))
df = df.sample(frac=1).reset_index(drop=True)
df
```

```python
#COnvert latin to utf-8
```

```python
type_int_list = ['views', 'likes', 'dislikes', 'comment_count']
for column in type_int_list:
    df[column] = df[column].astype(int)

type_str_list = ['category_id']
for column in type_str_list:
    df[column] = df[column].astype(str)
```

```python
#Set trending date to correct format of datetime
df['trending_date'] = pd.to_datetime(df['trending_date'], format='%y.%d.%m')
df['trending_date'].head()
```

```python
#Set publish time to correct format of datetime
df['publish_time'] = pd.to_datetime(df['publish_time'], format='%Y-%m-%dT%H:%M:%S.%fZ')
df['publish_time'].head()
```

```python
# separates'publish_time' column into 2 parts - 'publish_date' & 'publish_time' 
df.insert(4, 'publish_date',df['publish_time'].dt.date)
df['publish_time'] = df['publish_time'].dt.time
df[['publish_date', 'publish_time']].head()
```

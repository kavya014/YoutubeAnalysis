# Comparing No. of Instances of each dataset 

```python
size=[]
select = [data_CA , data_DE , data_FR , data_GB , data_IN , data_JP , data_KR , data_MX , 
            data_RU , data_US]
countries = ['Great Britan', 'India', 'Russia', 'Mexico', 'Japan',
             'United States', 'France', 'Denmark', 'South Korea', 'Canada']
for i in select:
    total_rows = len(i.index)
    size.append(total_rows)
```

```python
K=[]
for i in range(0,len(countries)):
    K.append(i)

plt.plot(K,size)
plt.xlabel('Countries' , fontsize =25)
plt.ylabel('No. of Instances',fontsize=25)
plt.xticks(K,countries,rotation=30,horizontalalignment='right')
plt.show()
```

# Comparing No. of Views

```python
views=[]
for i in select:
    temp = i['views'].mean()
    views.append(temp)
```

```python
plt.plot(K,views)
plt.xlabel('Countries' , fontsize =25)
plt.ylabel('No. of Views',fontsize=25)
plt.xticks(K,countries,rotation=30,horizontalalignment='right')
plt.show()
```

# Comparing No. of Likes

```python
likes=[]
for i in select:
    temp = i['likes'].mean()
    likes.append(temp)
    
plt.plot(K,likes)
plt.xlabel('Countries' , fontsize =25)
plt.ylabel('No. of Likes',fontsize=25)
plt.xticks(K,countries,rotation=30,horizontalalignment='right')
plt.show()
```

# Comparing No. of Dislikes

```python
dislikes=[]
for i in select:
    temp = i['dislikes'].mean()
    dislikes.append(temp)
    
plt.plot(K,dislikes)
plt.xlabel('Countries' , fontsize =25)
plt.ylabel('No. of Dislikes',fontsize=25)
plt.xticks(K,countries,rotation=30,horizontalalignment='right')
plt.show()
```

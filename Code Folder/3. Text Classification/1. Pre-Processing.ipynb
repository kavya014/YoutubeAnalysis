```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

import nltk
```

```python
data = pd.read_csv('YoutubeData.csv')
data.head()
```

# Pre-Processing Text

```python
#Normalizing case
data['title'] = data['title'].str.lower()
```

```python
#Tokenizer`
tokenized_title = []
from nltk.tokenize import word_tokenize
for i in range(len(data)) : 
    tokenized_word = word_tokenize(data.loc[i,'title'])
    tokenized_title.append(tokenized_word)
data['tokenized_title'] = tokenized_title
```

```python
#Removing stopwords
title_without_stopwords = []
from nltk.corpus import stopwords
stop_words=set(stopwords.words())
for i in range(len(data)) :
    tokenized_word = data.loc[i,'tokenized_title']
    tokens_without_sw = [word for word in tokenized_word if not word in stop_words]
    title_without_stopwords.append(tokens_without_sw)
data['title_without_stopwords'] = title_without_stopwords
```

```python
#Removing tokens that are not alphabetic
title_without_punc = []
for i in range(len(data)) :
    tokens = data.loc[i,'title_without_stopwords']
    words = [word for word in tokens if word.isalpha()]
    filtered_sentence = (" ").join(words)  #conversion to sentences
    title_without_punc.append(filtered_sentence) 
data['title_cleaned'] = title_without_punc
```

```python
#Transferring processed data to new csv file - data_cleaned.csv
na_rep = 'NULL'
na_rep_wrk = 8 * na_rep
data.to_csv("data_cleaned.csv", encoding="utf-8", index=False, na_rep=na_rep_wrk)
```

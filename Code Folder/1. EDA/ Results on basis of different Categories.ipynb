#3- Results on basis of different Categories

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#loading the dataset
data=pd.read_csv('YoutubeData.csv')
data.head()
```

```python
data = data[data['category_name'].notna()]
```

```python
unique_regions=pd.unique(data['region'])
```

```python
unique_categories=pd.unique(data['category_name'])
```

```python
labels = unique_categories
#size=[]
for x in unique_regions:
    size=[]
    for y in unique_categories:
        rslt_df = data[data['region'] == x]
        rslt_df=rslt_df[rslt_df['category_name']==y]
        size.append(rslt_df.shape[0])
   
    
    fig = plt.figure()
    ax = fig.add_axes([0,0,1,1])
    plt.title(x)
    plt.xticks(rotation=90)
    ax.bar(unique_categories,size)
    plt.show()
```

```python
K=[]
for i in range(0,len(unique_regions)):
    K.append(i)
listoflists=[]
c=[]
list=[]
for y in unique_categories:
    temp=[]
    for x in unique_regions:
        rslt_df = data[data['region'] == x]
        rslt_df=rslt_df[rslt_df['category_name']==y]
        temp.append(rslt_df.shape[0])
    c.append(y)
    listoflists.append(temp)

plt.plot(K,listoflists[0],label=c[0])
plt.plot(K,listoflists[1],label=c[1])
plt.plot(K,listoflists[2],label=c[2])
plt.plot(K,listoflists[3],label=c[3])
plt.plot(K,listoflists[4],label=c[4])
plt.plot(K,listoflists[5],label=c[5])

plt.xlabel('Countries',fontsize=15)
plt.ylabel('Count of videos',fontsize=15)
plt.title('Number of videos in each category in different countries',fontsize=20)
plt.xticks(K,unique_regions,rotation=30, horizontalalignment='right')
plt.legend(bbox_to_anchor=(1.0, 1.0))
plt.show()
```
